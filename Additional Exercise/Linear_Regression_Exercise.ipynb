```python
%%time
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from linreg import LinearModel
```

```python
#Explicitly specified the xlrd engine
df = pd.read_excel('chirps.xls', engine='xlrd')
df.head()
```

```python
x = df.X.values
y = df.Y.values
x = np.reshape(x, (x.shape[0], 1))
y = np.reshape(y, (y.shape[0], 1))
```

```python
model = LinearModel(1)

losses = model.train(
    x, y,
    200, 0.0001
)
```

```python
plt.plot(losses);
```

```python
y_preds = model.forward_pass(x)
```

```python
plt.plot(y_preds, y, 'r.');
```

```python
print(model.W, model.b)
```

### y = (4.748 * x) + 0.468

- This is the relationship that exists between **x (Number of chirps in a minute)** and **y (Ambient temperature of the room)**, based on the 15 examples in the given dataset.
